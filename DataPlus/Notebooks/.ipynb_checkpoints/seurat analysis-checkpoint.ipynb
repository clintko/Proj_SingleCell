# Read in data files

````r
dat_expr <- sapply(filenames, function(fname){
    
    dat <- read.table(file.path(fname), sep = "\t", header = TRUE)
    
    dat <- dat %>% 
        `rownames<-`(.[,"GENE"]) %>% 
        select(-GENE)
    
    return(dat)
})
````

# Exploratory Data Analysis (EDA)

````r
dat_sum = lapply(dat_expr, function(dat){
    tmp = data.frame(
        nUMI = apply(dat, 2, sum),
        nGene = apply(dat, 2, function(x){sum(x > 0.5)}))
    return(tmp)
})
````

````r
gp <- lapply(dat_sum, function(dat){
    tmp <- dat %>%
        ggplot(aes(x = nUMI / 1000)) +
        geom_histogram(binwidth = 1) +
        labs(x = "UMI (K)")
    return(tmp)
}) # end apply

options(repr.plot.width = 6, repr.plot.height = 6)
plot_grid(gp$Gland_10x, gp$Airway_10x, 
          gp$Gland_homeo, gp$Surface_naph, 
          labels = c("A (Yoshi)", "B (Yoshi)", "C (DropseqPipe)", "D (DropseqPipe)"),
          nrow = 2, ncol = 2, scale = 0.8)
````

````r
gp <- lapply(dat_sum, function(dat){
    tmp <- dat %>%
        ggplot(aes(x = nUMI / 1000, y = nGene / 1000)) + 
        geom_point(size = 0.5) +
        labs(x = "nUMI (K)", y = "nGene (K)")
    return(tmp)
}) # end apply

options(repr.plot.width = 6, repr.plot.height = 6)
plot_grid(gp$Gland_10x, gp$Airway_10x, 
          gp$Gland_homeo, gp$Surface_naph, 
          labels = c("A (Yoshi)", "B (Yoshi)", "C (DropseqPipe)", "D (DropseqPipe)"),
          nrow = 2, ncol = 2, scale = 0.8)
````

# Import to Create Seurat Object

```
CreateSeuratObject(raw.data, project = "SeuratProject", 
    min.cells = 0, min.genes = 0, is.expr = 0, 
    normalization.method = NULL, scale.factor = 10000, 
    do.scale = FALSE, do.center = FALSE,
    names.field = 1, names.delim = "_", meta.data = NULL,
    display.progress = TRUE, ...)
```

````r
# "Gland_10x", "Airway_10x", "Gland_homeo", "Surface_naph"
dat_seurat <- list()

dat_seurat$Gland_10x    <- CreateSeuratObject(
    raw.data  = dat_expr$Gland_10x, 
    project   = "Gland_10x", 
    min.cells = 1,
    min.genes = 500)

dat_seurat$Airway_10x   <- CreateSeuratObject(
    raw.data  = dat_expr$Airway_10x, 
    project   = "Airway_10x", 
    min.cells = 1,
    min.genes = 500)

dat_seurat$Gland_homeo  <- CreateSeuratObject(
    raw.data  = dat_expr$Gland_homeo, 
    project   = "Gland_homeo", 
    min.cells = 1,
    min.genes = 500)

dat_seurat$Surface_naph <- CreateSeuratObject(
    raw.data  = dat_expr$Surface_naph, 
    project   = "Surface_naph", 
    min.cells = 1,
    min.genes = 500)
````

filtering

````r
dat_seurat_filter <- lapply(dat_seurat, function(x){
    tmp <- FilterCells(
        object = x, 
        subset.names = c("nGene", "nUMI"), 
        low.thresholds  = c(200, -Inf), 
        high.thresholds = c(4000, 20000))
    return(tmp)
})
````

````r
options(repr.plot.width = 6, repr.plot.height = 6)
par(mfrow = c(2, 2))
for (idx in 1:4){
    GenePlot(object = dat_seurat_filter[[idx]], 
             gene1 = "nUMI", gene2 = "nGene", cex.use = 0.5)
}
par(mfrow = c(1, 1))
````

Normalization

````r
dat_seurat_norm <- lapply(dat_seurat_filter, function(x){
    tmp <- NormalizeData(
        object = x, 
        normalization.method = "LogNormalize", 
        scale.factor = 10000)
    return(tmp)
})
````

find variable genes

```
FindVariableGenes(object, mean.function = ExpMean,
    dispersion.function = LogVMR, do.plot = TRUE, set.var.genes = TRUE,
    x.low.cutoff = 0.1, x.high.cutoff = 8, y.cutoff = 1,
    y.high.cutoff = Inf, num.bin = 20, binning.method = "equal_width",
    selection.method = "mean.var.plot", top.genes = 1000, do.recalc = TRUE,
    sort.results = TRUE, do.cpp = TRUE, display.progress = TRUE, ...)
```

````r
dat_seurat_FVG <- lapply(dat_seurat_norm, function(x){
    tmp <- FindVariableGenes(
        object = x, 
        mean.function = ExpMean, 
        dispersion.function = LogVMR, 
        x.low.cutoff = 0.0125, 
        x.high.cutoff = 3, 
        y.cutoff = 0.5, 
        do.plot = FALSE)
    return(tmp)
})
````

````r
options(repr.plot.width = 7, repr.plot.height = 7)
par(mfrow = c(2, 2))
for (idx in 1:4){
    VariableGenePlot(
        object = dat_seurat_FVG[[idx]], 
        cex.text.use = 0.7,
        cex.use = 0.3)
}
par(mfrow = c(1, 1))
````

### Scaled

````r
dat_seurat_scaled <- lapply(dat_seurat_FVG, function(x){
    ScaleData(object = x)
})
````

# Run PCA

````r
dat_seurat_PCA <- lapply(dat_seurat_scaled, function(x){
    tmp = RunPCA(object = x, do.print = FALSE)
    return(tmp)
})
````

````r
options(repr.plot.width = 4, repr.plot.height = 2.5)
par(mfrow = c(2, 2))
for (idx in 1:4){
    PCAPlot(object = dat_seurat_PCA[[idx]], dim.1 = 1, dim.2 = 2)
}
par(mfrow = c(1, 1))
````

# Find Clusters

````r
dat_seurat_Clust <- lapply(dat_seurat_PCA, function(x){
    tmp = FindClusters(
        object = x, 
        reduction.type = "pca", 
        dims.use = 1:10, 
        resolution = 0.6, 
        print.output = 0, 
        save.SNN = TRUE)
    return(tmp)
})
````

# T-SNE

````r
dat_seurat_tsne <- lapply(dat_seurat_Clust, function(x){
    tmp = RunTSNE(
        object = x, 
        reduction.use = "pca",
        dims.use = 1:10, 
        do.fast = TRUE)
    return(tmp)
})
````

````r
options(repr.plot.width = 4, repr.plot.height = 2.5)
par(mfrow = c(2, 2))
for (idx in 1:4){
    TSNEPlot(object = dat_seurat_tsne[[idx]])
}
par(mfrow = c(1, 1))
````
