```python
fig, axes = plt.subplots(1, 3, figsize = (12, 4))
#axes.scatter(subdata[:, 0], subdata[:, 1], s = 3, alpha = 0.8)
#axes.set(xscale="log", yscale="log")

subdata_name = ["Costim", "CMV", "SEB"]
for idx_data in range(3):
    # intialization
    ax = axes[idx_data]
    data = subdata[idx_data]
    title = subdata_name[idx_data]
    
    # plot scatter plot
    sns.regplot(
        data[:, 0], data[:, 1], 
        ax = ax,
        fit_reg = False, 
        scatter_kws={'s': 5})
    
    # set plot
    ax.set_title(title)
    ax.set_xlabel("CD4")
    ax.set_ylabel("CD8")

plt.tight_layout()
plt.show()
```

# Plot different combination of variables

```python
def plot_scatter(data, ax, var1, var2, k = 1000):
    """plot scatter plot a pair of variables"""
    
    # which colnum
    idx_marker = [markers_dict[var1], markers_dict[var2]] 
    # which row
    if k == -1:
        idx_sample  = range(data.shape[0])
    else:
        idx_sample  = rand.permutation(range(data.shape[0]))[:k]             
    # combine row, column
    idx = np.ix_(idx_sample, idx_marker)
    
    # get subdata
    subdata = data[idx]
    #print(idx_marker)
    #print(subdata.shape)
    
    
    # plot scatter plot
    sns.regplot(
        subdata[:, 0], subdata[:, 1], 
        ax = ax,
        fit_reg = False, 
        scatter_kws={'s': 5})
    
    # set plot
    ax.set_xlabel(var1)
    ax.set_ylabel(var2)
```

```python
markers_dict
```

```python
var1 = ['FSC-H', 'CD3 AmCyan',       'CD4 PerCP Cy55', 'CD45RO ECD', 'CD45RO ECD', 'CD4 PerCP Cy55', 'CD4 PerCP Cy55']
var2 = ['FSC-W', 'CD14 CD19 vAmine', 'CD8 APC Cy7',    'CD27 APC',   'CD57 FITC',  'TNFa A700', 'IL 2 PE']
```

```python
len(var1)
```

```python
print(var1[0], markers_dict[var1[0]])
print(var2[0], markers_dict[var2[0]])

fig, ax = plt.subplots(1, 1, figsize = (4, 4))
plot_scatter(data_all[1], ax, var1[0], var2[0], k = -1)
plt.show()
```

```python
print(var1[1], markers_dict[var1[1]])
print(var2[1], markers_dict[var2[1]])

fig, ax = plt.subplots(1, 1, figsize = (4, 4))
plot_scatter(data_all[1], ax, var1[1], var2[1], k = -1)
plt.show()
```

```python
fig, axes = plt.subplots(len(var1), 3, figsize = (4 * 3, 4 * len(var1)))

# set title
subdata_name = ["Costim", "CMV", "SEB"]
for idx in range(3):
    title = subdata_name[idx]
    ax = axes[0, idx]
    ax.set_title(title)

for idx_var in range(len(var1)):    
    for idx_data in range(3):
        ax = axes[idx_var, idx_data]
        data = data_all[idx_data]
        plot_scatter(data, ax, var1[idx_var], var2[idx_var], k = 10000)
    
plt.tight_layout()
plt.show()
```

# Converted to .np files

```python
import glob
```

```python
# flow files
data_dir = "/data/flow/EQAPOL_normal/"
files = [glob.glob(data_dir + "*" + idx_name + ".txt") 
         for idx_name in ['Costim', 'CMV_pp65', 'SEB']]
files = np.array(files)

# makers
markers = np.loadtxt(data_dir + 'markers.txt', dtype=str, delimiter=",")
markers_dict = {items[1]: idx for idx, items in enumerate(markers)}

def read_data(file):
    '''the function print the file it is reading in; a numpy array is returned'''
    print(file)
    return np.loadtxt(file)

print("Read in Costim data")
data_Costim = [np.loadtxt(file) for file in files[0]]
print("Read in CMV data")
#data_CMV    = [read_data(file) for file in files[1]]
data_CMV    = [np.loadtxt(file) for file in files[1]]
print("Read in SEB data")
data_SEB    = [np.loadtxt(file)  for file in files[2]]


data_Costim = np.array(data_Costim)
data_CMV    = np.array(data_CMV)
data_SEB    = np.array(data_SEB)
```

```python
# output the value
tmp_dir = "/data/deep-learning-group/test_data/flow_EQAPOL/"
fname_Costim = "data_Costim.np"
fname_CMV    = "data_CMV.np"
fname_SEB    = "data_SEB.np"

# open and write the file
print("Store Costim")
file_object = open(tmp_dir + fname_Costim, 'wb')
np.save(file_object, data_Costim)
file_object.close()

print("Store CMV")
file_object = open(tmp_dir + fname_CMV, 'wb')
np.save(file_object, data_CMV)
file_object.close()

print("Store SEB")
file_object = open(tmp_dir + fname_SEB, 'wb')
np.save(file_object, data_SEB)
file_object.close()

print("Store markers")
file_object = open(tmp_dir + "markers.np", 'wb')
np.save(file_object, markers)
file_object.close()

print("The data are stored.")
```
