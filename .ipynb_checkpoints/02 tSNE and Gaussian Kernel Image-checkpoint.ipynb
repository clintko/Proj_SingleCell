# read in data

```python
# tools
import numpy as np
import numpy.random as rand
import scipy as sp
from scipy.interpolate import Rbf
import pandas as pd
from sklearn.manifold import TSNE

# data IO
import pickle

# plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ls
```

```python
file_name = "data_simulation"
file_object = open(file_name, 'rb')  
samples = pickle.load(file_object) 
file_object.close()
```

```python
# check dimension
samples.shape
```

```python
m, n, p = samples.shape
```

# standardization

```python
tmp = np.vstack(samples)
tmp.shape
tmp[0:6]
```

```python
#scale_mu = np.apply_along_axis(np.mean, 0, tmp)
#print(scale_mu)
#scale_sd = np.apply_along_axis(np.std,   0, tmp)
#print(scale_sd)

# standardization
tmp = np.vstack(samples)
tmp = (tmp - np.mean(tmp, axis=0)) / np.std(tmp, axis=0)

# check dimension
print(tmp.shape)

# reshape the matrix back
tmp = np.expand_dims(tmp, axis = 0)
tmp = tmp.reshape(m, n, p)
print(tmp.shape)

samples_stdard = tmp
```

# Generate t-SNE plot

Note:   
https://medium.com/@luckylwk/visualising-high-dimensional-datasets-using-pca-and-t-sne-in-python-8ef87e7915b


```python
data = samples_stdard[0]
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(data)
```

```python
tsne_results.shape
```

```python
data = samples_stdard[0]
tmp = tsne_results

fig, axes = plt.subplots(2, 2, figsize=(7,7))
axes = axes.ravel()

color = ["#fbb4ae"] * 903 + ["#b3cde3"] * 97
axes[0].scatter(tmp[:, 0], tmp[:, 1], s=10, alpha=1, color = color)
axes[0].title.set_text("label")

z = data[:, 0]
axes[1].scatter(
    tmp[:, 0], tmp[:, 1], 
    s=10, alpha=1, c = z, 
    cmap=plt.get_cmap('Reds'))
axes[1].title.set_text("gene 1")

z = data[:, 1]
axes[2].scatter(
    tmp[:, 0], tmp[:, 1], 
    s=10, alpha=1, c = z, 
    cmap=plt.get_cmap('Reds'))
axes[2].title.set_text("gene 2")

z = data[:, 2]
axes[3].scatter(
    tmp[:, 0], tmp[:, 1], 
    s=10, alpha=1, c = z, 
    cmap=plt.get_cmap('Reds'))
axes[3].title.set_text("gene 3")

plt.show()
```

# convert t-SNE plot into image

initialization

```python
n_grid = 50
```

```python
tmp = tsne_results
x = tmp[:, 0]
x_c = np.linspace(min(x), max(x), n_grid)

y = tmp[:, 1]
y_c = np.linspace(min(y), max(y), n_grid)

x_c, y_c = np.meshgrid(x_c, y_c)
x_c = x_c.ravel()
y_c = y_c.ravel()
```

```python
# https://docs.scipy.org/doc/scipy-0.16.1/reference/generated/scipy.interpolate.Rbf.html
# https://docs.scipy.org/doc/scipy/reference/tutorial/interpolate.html

data = samples_stdard[0]
z = data[:, 0]

# gaussian interpolation using radial basis function
rbfi = Rbf(x, y, z, function='gaussian', epsilon = 1)
z_c = rbfi(x_c, y_c)

# linear scaling
z_c = (z_c - min(z_c)) / (max(z_c) - min(z_c))
```

```python
data = samples_stdard[0]
tmp = tsne_results
z = data[:, 0]

#color = ["#fbb4ae"] * 903 + ["#b3cde3"] * 97
#label
fig, axes = plt.subplots(1, 1, figsize=(7,7))

axes.scatter(
    x_c, y_c, 
    s=10, alpha=1, c = z_c, 
    cmap=plt.get_cmap('Blues'))

axes.scatter(
    tmp[:, 0], tmp[:, 1], 
    s=10, alpha=1, c = z, 
    cmap=plt.get_cmap('Reds'))

plt.show()
```
